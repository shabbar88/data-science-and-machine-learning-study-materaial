```python
def cost_budget(price):
    if price<10:
        return '*'
    elif price>10 and price<20:
        return '**'
    elif price>20:
        return '***'
```

```python
# function inside apply should return single value
dataframe['Cost_Budget']=dataframe['total_bill'].apply(cost_budget)
```

```python
dataframe
```

```python
def quality(total_bill,tip):
    if tip/total_bill>0.25:
        return "Generous"
    else:
        return "Other"
```

```python
quality(1600,160)
```

```python
# this is the way we can combine lamdda and apply in pandas
dataframe['Quality']=dataframe[['total_bill','tip']].apply(lambda dataframe:quality(dataframe['total_bill'],dataframe['tip']),axis=1)
```

```python
# merge()
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations,logins,how='inner',on='name')
```

```python
pd.merge(registrations,logins,how='outer',on='name')
```

```python
pd.merge(registrations,logins,how='left',on='name')
```

```python
pd.merge(registrations,logins,how='right',on='name')
```
