```python
                       # count plot
    # import libraries
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns 
    
```

```python
# countplot is a specvial case of bar plot
# it allows to use any sort of  a matrix
plt.figure(figsize=(10,4))
sns.countplot(data=data,x='division',hue='level of education',palette='Dark2')
```

```python
# boxplot
# boxplot displays distribution of a continiuous variable
# it does this through thbe use of a quartile
# quartile seperates out into four number of data points
# Q1,Q2,Q3,Q4 ARE THE FOUR QUARTILES
# 25 % OF DATA IN BOTTOM QUARTILES
# Q2 IS 50TH PERCENTILE OR MEDIAN
# IQR=Q3-Q1 I.E INTERQUARTILE ANGE
#1.DATA WHICH IS BEYOND THE BOUNDARRY OF 1.5 IQR IS CALLED AS OUTLIERS
# WE CAN CREATE ONE BOXPLOT PER CATEGORY


# VIOLIN PLOT PLAYS A SIMILAR ROLE TO BOXPLOT
# IT DISPLAYS PROBABILITY DENSITY CURVE  ACROSS DATA USIMG KDE
# VIOLIN PLOT CAN BE CRFEATED AS PER CATEGORY
```

```python
    # import libraries
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
mydata=pd.read_csv('StudentsPerformance.csv')
```

```python
mydata
```

```python
mydata['test preparation course'].value_counts()
```

```python
mydata['test preparation course']=mydata['test preparation course'].map({'none':0,'completed':1})
```

```python
mydata['test preparation course'].value_counts()
```

```python
sns.boxplot(data=mydata,y='math score',x ='test preparation course')
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(data=mydata,y='math score',x ='test preparation course',hue='parental level of education')
plt.legend(bbox_to_anchor=(1.0,0.5))
```

```python
# violinplot
plt.figure(figsize=(10,7))
sns.violinplot(data=mydata,y='math score',x ='test preparation course',hue='parental level of education')
plt.legend(bbox_to_anchor=(1.0,0.5))
```

```python
# heatmap()
# wer should have all cells in the same unit so the color mapping makes sense 

df = pd.read_csv('country_table.csv')
```

```python
plt.figure(figsize=(10,6),dpi=150)
sns.heatmap(df.drop('Life expectancy',axis=1),annot=True,cmap='viridis')
plt.tight_layout()
plt.savefig('heatmap1111.png')
```
