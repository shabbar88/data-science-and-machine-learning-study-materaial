```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
heart=pd.read_csv('heart-disease.csv')
```

```python
heart.head()
```

```python
over_50=heart[heart['age']>50]
```

```python
over_50
```

```python
fig,ax=plt.subplots(figsize=(4,6),dpi=200,nrows=2,ncols=1)
scatter=ax[0].scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'],cmap='winter')
# customising the plot
ax[0].set(title="Heart Disease And Cholestorol",ylabel="Cholestorol",xlabel='age')
ax[0].legend(*scatter.legend_elements(),title="target")
fig.savefig("heart-dis.png",bbox_inches="tight")
# adding horizontal line
ax[0].axhline(y=over_50['chol'].mean(), color='red', linestyle='--',lw=2)


# now we will plot age vs thalac
scatter=ax[1].scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'],cmap='summer')
# customising the plot
ax[1].set(title="Heart Disease And Thalach",xlabel="Age",ylabel="Thalach")
ax[1].legend(*scatter.legend_elements(),title="target")
fig.savefig("heart-dis.png",bbox_inches="tight")
# adding horizontal line
ax[1].axhline(y=over_50['thalach'].mean(), color='red', linestyle='--',lw=2)

fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold")
fig.tight_layout()
```
